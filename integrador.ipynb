# Critérios de pré-processamento:
- Dados nulos serão excluídos
- Serão excluídas as colunas que não agregam ao objetivo



#### Verificando se há dados nulos e ecluíndo

```python
def pre_processamento(df_origem, excluir_colunas=None):
  if df_origem.isnull().values.any():
    df_origem = df_origem.dropna()

  if excluir_colunas:
    df_origem = df_origem.drop(columns=excluir_colunas)

  return df_origem
```

```python
df_origem = pre_processamento(df_origem, excluir_colunas=['product_id', 'discounted_price', 'discount_percentage', 'rating_count', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title',
       'review_content', 'img_link', 'product_link'])
```

#### Salvando os dados em uma tabela.csv com os dados processados

```python
df_origem.to_csv('df_dados_processados.csv', index=False)
```

```python
df_processados = pd.read_csv('https://raw.githubusercontent.com/AlexandreP23/Amazon-Sales-Dataset/main/DS_Amazon/df_dados_processados.csv')
df_processados
```

#### Gerando críticas
- Avaliações menores que 4.0 não serão disponibilizadas

```python
def criticas(df_processados):
    df_processados['rating'] = pd.to_numeric(df_processados['rating'], errors='coerce')
    df_criticados = df_processados[df_processados['rating'] > 4.0]
    return df_criticados
```

```python
df_criticados = criticas(df_processados)
df_criticados
```

#### Salvando os dados em uma tabela.csv com os dados criticados

```python
df_criticados.to_csv('df_dados_criticados.csv', index=False)
```
